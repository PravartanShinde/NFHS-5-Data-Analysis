# NFHS-5 Data Cleaning

This notebook documents the data cleaning and preprocessing steps performed on the National Family Health Survey (NFHS-5) district-level dataset.

The NFHS-5 survey was conducted across India during **2019–2021** under the stewardship of the International Institute for Population Sciences (IIPS), Mumbai, with technical support from the ICF. The survey provides comprehensive information on population, health, nutrition, and socioeconomic indicators at the national, state, and district levels.

The objective of this notebook is to prepare the dataset for exploratory data analysis by identifying and addressing data quality issues. The cleaning workflow includes:

* Initial dataset inspection and validation
* Column name standardization
* Duplicate record detection
* Missing value analysis
* Handling variables with excessive missing values
* Missing value imputation
* Final dataset validation and export

The resulting cleaned dataset serves as the foundation for subsequent analyses on lifestyle and nutrition indicators across Indian states and districts.

## Cloned data set into df_cleann

```python
df_cleann = df.copy()   ## clean data 
```

```python
df_cleann.columns = (
    df_cleann.columns
    .str.replace(r"\(.*?\)", "", regex=True)
    .str.replace(",", "")
    .str.replace("%", "")
    .str.replace("-", " ")
    .str.replace(r"[^\w\s]", "", regex=True)
    .str.strip()
)

df_cleann.columns = (
    df_cleann.columns
    .str.lower()
    .str.replace(r"\s+", "_", regex=True)
    .str.replace(r"_?scaling_factor\d+", "", regex=True)
)
```

```python
df_cleann.columns
```

## Column names were standardized to improve readability and ensure consistency throughout the analysis. Special characters, percentage symbols, brackets, and unnecessary text were removed, while all column names were converted to lowercase and formatted using underscores.

## Missing values Analysis
This section examines the extent and distribution of missing values in the dataset. Identifying missing data is an important step in the cleaning process, as it helps determine whether variables should be removed, imputed, or retained for further analysis.

```python
df_cleann.isnull().sum().sort_values(ascending=False)
```

### Observation

The missing value analysis revealed that most variables contain little to no missing data. However, a small number of child nutrition and healthcare-related indicators exhibited substantial missingness, with the highest missing count reaching 329 observations. These variables were examined separately to determine whether they should be retained, imputed, or removed from the dataset.

```python
missing_percent = df_cleann.isnull().mean() * 100
```

```python
missing_percent[missing_percent > 0].sort_values()
```

### Observation

The majority of variables contain very little missing data, with most columns having less than 1% missing values. However, a small number of child nutrition and healthcare-related indicators exhibit substantial missingness. The highest missing percentages were observed in:

- Children aged 6–8 months receiving solid or semi-solid food and breastmilk (46.7%)
- Non-breastfeeding children aged 6–23 months receiving an adequate diet (46.6%)
- Several diarrhoea-related indicators (~36.9%)

Given the high proportion of missing values in these variables, they require special consideration before further analysis.

```python
df_cleann["children_age_group_6_to_8_months_receiving_solid_or_semisolid_food_and_breastmilk"].describe()
```

```python
col = "children_age_group_6_to_8_months_receiving_solid_or_semisolid_food_and_breastmilk"

nan_count = df_cleann[col].isnull().sum()

zero_count = (df_cleann[col] == 0).sum()

nan_count, zero_count
```

```python
df_cleann = df_cleann.drop(
    columns=[
        "children_age_group_6_to_8_months_receiving_solid_or_semisolid_food_and_breastmilk"
    ]
)
```

### Observation

The variable `children_age_group_6_to_8_months_receiving_solid_or_semisolid_food_and_breastmilk` contained 46.7% missing values (329 observations). In addition, 311 of the remaining observations were recorded as zero, resulting in over 90% of the dataset containing either missing or zero values for this variable. Due to the high level of missingness and limited informational value, the column was removed from the dataset.

```python
#df_cleann.shape ## checking if the column got dropped 
```

```python
col = "non_breastfeeding_children_age_group_6_to_23_months_receiving_an_adequate_diet"

nan_count = df_cleann[col].isnull().sum()

zero_count = (df_cleann[col] == 0).sum()

non_zero_count = (df_cleann[col] > 0).sum()

nan_count, zero_count, non_zero_count
```

```python
df_cleann = df_cleann.drop(
    columns=[
        "non_breastfeeding_children_age_group_6_to_23_months_receiving_an_adequate_diet"
    ]
)
```

### Observation

The variable `non_breastfeeding_children_age_group_6_to_23_months_receiving_an_adequate_diet` contained 328 missing values (46.6% of observations). Additionally, 317 observations recorded a value of zero, leaving only 59 non-zero observations across the dataset. Since more than 91% of the records contained either missing or zero values, the variable was removed due to its limited analytical value.

### Choosing an Imputation Strategy

To determine an appropriate imputation method, the skewness of variables containing missing values was examined. Several variables exhibited substantial positive or negative skewness, indicating non-symmetric distributions. Therefore, median imputation was selected because it is more robust to skewed data and extreme values than mean imputation.

```python
df_clean = df_clear.fillna(df_clear.median(numeric_only=True))
```
